```python
import pandas as pd
import matplotlib.pyplot as plt

# Laad de EV dataset
ev_data = pd.read_csv('IEA-EV-dataEV salesCarsHistorical.csv')

# Data voor EV verkoop voorbereiden
ev_sales = ev_data.groupby('year')['value'].sum().reset_index()

# Plot voor EV verkoop
plt.figure(figsize=(12, 6))
plt.plot(ev_sales['year'], ev_sales['value'], marker='o', linestyle='-', color='b')
plt.xlabel('Jaar')
plt.ylabel('Verkochte Aantallen')
plt.title('Verkooptrend van Elektrische Voertuigen door de Jaren Heen')
plt.grid(True)
plt.show()
```

```python
# Importeren van de benodigde bibliotheken
import pandas as pd
import matplotlib.pyplot as plt

# Laad de dataset
co2_data = pd.read_csv('co2_emissions_kt_by_country.csv')

# Bekijk de eerste paar rijen van de dataset
print(co2_data.head())

# Controleer de unieke jaren in de dataset
unique_years = co2_data['year'].unique()
unique_years.sort()
print(unique_years)

# Groepeer de data op jaar en bereken de totale CO2-uitstoot per jaar
total_co2_per_year = co2_data.groupby('year')['value'].sum().reset_index()

# Maak de plot
plt.figure(figsize=(14, 8))
plt.plot(total_co2_per_year['year'], total_co2_per_year['value'], marker='o', linestyle='-', color='b')
plt.xlabel('Jaar')
plt.ylabel('CO2-uitstoot (kiloton)')
plt.title('Totale CO2-uitstoot door de jaren heen')
plt.grid(True)
plt.show()

```
