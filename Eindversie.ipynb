## Positieve Impact van EVs op de Luchtkwaliteit


Elektrische voertuigen hebben een aanzienlijke positieve impact op de luchtkwaliteit. Door hun emissievrije werking dragen elektrische voertuigen bij aan schonere lucht en een gezondere leefomgeving. Landen die meer gebruik maken van elektrische voertuigen zullen minder CO2 uitstoten, waardoor de luchtkwaliteit in die landen beter zal zijn dan landen die minder gebruik maken van elektrische voertuigen.

### Verkoop van elektrische voertuigen

Het is een feit dat elektrische voertuigen tot 60% minder CO2 uitstoten dan voertuigen op benzine of diesel (rijksdienst: Milieu en elektrisch vervoer). Doordat er meer elektrische voertuigen worden verkocht, zullen er minder voertuigen op benzine of diesel rijden en zal de luchtkwaliteit verbeteren. 

```python
import pandas as pd
import matplotlib.pyplot as plt

# Laad de EV dataset
ev_data = pd.read_csv('IEA-EV-dataEV salesCarsHistorical.csv')

# Data voor EV verkoop voorbereiden
ev_sales = ev_data.groupby('year')['value'].sum().reset_index()

# Plot voor EV verkoop
plt.figure(figsize=(12, 6))
plt.plot(ev_sales['year'], ev_sales['value'], marker='o', linestyle='-', color='b')
plt.xlabel('Jaar')
plt.ylabel('Verkochte Aantallen')
plt.title('Verkooptrend van Elektrische Voertuigen door de Jaren Heen')
plt.grid(True)
plt.show()
```

> *Figure 1: X-as: staan de jaren van 2010 tot en met 2022. Y-as: staan de verkochte aantallen keer 10 miljoen.

Door te kijken naar de verkooptrend kunnen we zien dat de verkoop van elekrische voertuigen vanaf 2016 pas is gaan stijgen. We kunnen vaststellen dat door de groei van de verkoop van elektrische voertuigen er minder CO2 zal worden uitgestoten en er dus een betere luchtkwaliteit zal zijn. De groei van de verkooptrend zal de CO2-uitstoot laten dalen.

### Vergelijking van verkochte elektrische voertuigen en CO2-uitstoot per land.

Elektrische voertuigen spelen een cruciale rol in het verminderen van de CO2-uitstoot en het verbeteren van de luchtkwaliteit. Door de verkoopgegevens van EVs te vergelijken met de CO2-uitstoot in verschillende landen, kunnen we inzicht krijgen in de effectiviteit van EV-adoptie op het verminderen van emissies. Omdat we niet alle landen wilden doen, hebben we er 5 gekozen: Nederland, Spanje, Japan, Frankrijk en Canada. Door minder landen te gebruiken kan er makkelijker per land vergeleken worden. 

```python
co2_data = pd.read_csv('co2_emissions_kt_by_country.csv')
# Selecteer de landen die overeenkomen met de eerder getoonde EV verkoop grafiek
countries = ['Canada', 'France', 'Japan', 'Netherlands', 'Spain', ]

# Filter de CO2-uitstoot dataset voor deze landen
filtered_co2_data = co2_data[co2_data['country_name'].isin(countries)]

# Bereken de gemiddelde CO2-uitstoot voor deze landen
filtered_avg_co2 = filtered_co2_data.groupby('country_name')['value'].mean().reset_index()

import matplotlib.pyplot as plt

# Plot de gemiddelde CO2-uitstoot voor de geselecteerde landen
plt.figure(figsize=(6, 3))
plt.bar(filtered_avg_co2['country_name'], filtered_avg_co2['value'], color='red')
plt.xlabel('Land')
plt.ylabel('Gemiddelde CO2-uitstoot (kiloton)')
plt.title('Gemiddelde CO2-uitstoot voor geselecteerde landen met EV verkoop data')
plt.xticks(rotation=90)
plt.show()
```

> *Figure 2: X-as: staan de 5 gekozen landen. Y-as: staan de gemiddelde CO2 uitsoot in kiloton.

In figuur 2 zie je dat Canada en frankrijk een gemiddelde CO2-uitstoot hebben, Japan heeft een hoge CO2-uitstoot en Nederland en Spanje hebben een vrij lage CO2-uitstoot vergeleken elkaar. 

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the EV sales dataset
ev_sales_data = pd.read_csv('IEA-EV-dataEV salesCarsHistorical.csv')

# Filter data for the specified countries
filtered_countries = ['Netherlands', 'France', 'Canada', 'Japan', 'Spain']
filtered_ev_sales = ev_sales_data[ev_sales_data['region'].isin(filtered_countries)]

# Group the data by country and sum the sales
ev_sales_per_country = filtered_ev_sales.groupby('region')['value'].sum().reset_index()

# Plot the data
plt.figure(figsize=(6, 3))
plt.bar(ev_sales_per_country['region'], ev_sales_per_country['value'], color='green')
plt.xlabel('Land')
plt.ylabel('Verkoop van Elektrische Voertuigen')
plt.title('Totale Verkoop van Elektrische Voertuigen in Geselecteerde Landen')
plt.xticks(rotation=90)
plt.show()


```

> *Figure 3: X-as: staan de 5 gekozen landen. Y-as: staat de verkoop van elektrische voertuigen keer 1 miljoen.

In figuur 3 zie je dat vergeleken met elkaar Canada, Japan en Nederland een vrij gemiddelde verkoop hebben, Frankrijk een hele hoge en Spanje een lagere verkoop hebben. 

De vergelijking van de verkoop van elektrische voertuigen met de CO2-uitstoot in Nederland, Frankrijk, Canada, Japan en Spanje laat interessante inzichten zien in de effectiviteit van de adoptie van elektrische voertuigen op de vermindering van CO2-uitstoot. In Canada zien we dat er een gemiddelde verkoop van elektrische voertuigen is en een gemiddelde CO2-uitstoot is, dit laat niet zien dat elektrische voertuigen impact hebben op de vermindering van CO2-uitstoot. Frankrijk daarentegen toont een relatief hoge verkoop van EVs gecombineerd met een lage gemiddelde CO2-uitstoot, wat de positieve impact van elektrische voertuigen adoptie benadrukt. In Japan blijft de gemiddelde CO2-uitstoot hoog ondanks een gemiddelde verkoop van elektrische voertuigen, dit laat niet zien dat elektrische voertuigen impact hebben op de vermindering van CO2-uitstoot. Nederland vertoont een hoge verkoop van EVs en een lage CO2-uitstoot, wat de positieve impact van elektrische voertuigen adoptie benadrukt. Spanje heeft beide relatief laag, dit laat niet zien dat elektrische voertuigen impact hebben op de vermindering van CO2-uitstoot. Deze vergelijking toont aan dat de verkoop van elektrische voertuigen een belangrijke rol speelt in het verminderen van de CO2-uitstoot en het verbeteren van de luchtkwaliteit, de reden dat een aantal landen meer CO2-uitstoot hebben kan ook verklaard worden door andere bronnen van CO2-uitstoot. 

### verkoop elektrische voertuigen vergeleken de gemiddelde luchtkwaliteit van elk land.

Door de verkochte aantal elektrische voertuigen van elk land te vergelijken met de gemiddelde luchtkwaliteit van elk land, zal worden laten zien dat door de verkoop van elektrische voertuigen de luchtkwaliteit zal verbeteren. Dit zullen we laten zien door 2 datasets te vergelijken in een scatterplot.

```python
import pandas as pd

# Load the datasets
ev_sales_data = pd.read_csv('IEA-EV-dataEV salesCarsHistorical.csv')
co2_emissions_data = pd.read_csv('co2_emissions_kt_by_country.csv')
aqi_data = pd.read_csv('data_date.csv')

# Filter data for the specified countries
filtered_countries = ['Netherlands', 'France', 'Canada', 'Japan', 'Spain']
filtered_ev_sales = ev_sales_data[ev_sales_data['region'].isin(filtered_countries)]
filtered_co2_emissions = co2_emissions_data[co2_emissions_data['country_name'].isin(filtered_countries)]
filtered_aqi = aqi_data[aqi_data['Country'].isin(filtered_countries)]

# Group the EV sales data by country and sum the sales
ev_sales_per_country = filtered_ev_sales.groupby('region')['value'].sum().reset_index()

# Calculate average CO2 emissions per country over the years
avg_co2_emissions = filtered_co2_emissions.groupby('country_name')['value'].mean().reset_index()

# Calculate average AQI per country over the years
avg_aqi = filtered_aqi.groupby('Country')['AQI Value'].mean().reset_index()

# Merge datasets
merged_data = pd.merge(ev_sales_per_country, avg_co2_emissions, left_on='region', right_on='country_name')
merged_data = pd.merge(merged_data, avg_aqi, left_on='region', right_on='Country')
merged_data = merged_data.rename(columns={'value_x': 'EV Sales', 'value_y': 'Avg CO2 Emissions', 'AQI Value': 'Avg AQI'})


import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(14, 8))

# Scatter plot with regression line
sns.regplot(x='EV Sales', y='Avg AQI', data=merged_data, scatter_kws={'s':100}, color='red')
plt.title('EV Sales vs. Average AQI')
plt.xlabel('EV Sales')
plt.ylabel('Avg AQI')
plt.show()
```

>Figure 4: 
X-as: Toont het totale aantal verkochte elektrische voertuigen in elk geselecteerd land. 

>Y-as: Toont de gemiddelde luchtkwaliteitsindex (AQI) in elk land. Een lagere AQI-waarde betekent betere luchtkwaliteit.

>Punten: Elk rood punt in de plot vertegenwoordigt een specifiek land uit de geselecteerde landen (Nederland, Frankrijk, Canada, Japan en Spanje). De positie van elk punt geeft aan hoeveel elektrische voertuigen in dat land zijn verkocht (X-as) en wat de gemiddelde luchtkwaliteitsindex is (Y-as).

>Regressielijn: De rode lijn in de plot is een regressielijn die de algemene trend weergeeft tussen EV Sales en Average AQI. De lijn heeft een negatieve helling, wat betekent dat er een afname van de gemiddelde AQI is naarmate de verkoop van EVs toeneemt.

>Schaduwgebied (95% betrouwbaarheidsinterval): Het rode schaduwgebied rond de regressielijn geeft het 95% betrouwbaarheidsinterval weer. Dit interval toont de mate van onzekerheid rond de regressielijn. Een smaller schaduwgebied wijst op een hogere betrouwbaarheid van de regressielijn, terwijl een breder schaduwgebied duidt op meer variabiliteit en onzekerheid.



De scatterplot toont een negatieve correlatie tussen de verkoop van elektrische voertuigen en de gemiddelde luchtkwaliteitsindex (AQI). Dit suggereert dat landen met een hogere verkoop van elektrische voertuigen meestal een betere luchtkwaliteit hebben (lagere AQI). Elk punt in de plot vertegenwoordigt een specifiek land, en de regressielijn toont de algemene trend: naarmate de verkoop van elektrische voertuigen toeneemt, verbetert de luchtkwaliteit.

Door deze visualisaties kunnen we concluderen dat de adoptie van elektrische voertuigen een kleine rol speelt in het verbeteren van de luchtkwaliteit en het verminderen van de CO2-uitstoot.

## Beperkte Impact van EVs op de Luchtkwaliteit

De adoptie van elektrische voertuigen heeft een beperkte impact op de algehele luchtkwaliteit vanwege andere bestaande vervuilingsbronnen en het huidige marktaandeel van elektrische voertuigen. Dit perspectief wordt onderzocht door data die gebieden belicht waar, ondanks een toename van het gebruik van elektrische voertuigen, de luchtkwaliteit niet significant is verbeterd.

### Luchtkwaliteit blijft matig ondanks groei verkoop elektrische voertuigen

Zelfs met de stijgende verkoop van elektrische voertuigen, is de luchtkwaliteit in veel gebieden nog steeds slecht. Dit wijst erop dat er andere vervuilingsbronnen zijn die bijdragen aan luchtverontreiniging. Elektrische voertuigen hebben dus amper impact op de luchtkwaliteit van landen.



```python
import plotly.express as px
ev_sales_data = pd.read_csv('IEA-EV-dataEV salesCarsHistorical.csv')
co2_emissions_data = pd.read_csv('co2_emissions_kt_by_country.csv')
aqi_data = pd.read_csv('data_date.csv')


# Aggregating the air quality data by country
air_quality_agg = aqi_data.groupby('Country')['AQI Value'].mean().reset_index()
air_quality_agg.rename(columns={'Country': 'country_name', 'AQI Value': 'avg_aqi'}, inplace=True)

# Aggregating the EV sales data by country
ev_sales_total = ev_sales_data.groupby('region')['value'].sum().reset_index()
ev_sales_total.rename(columns={'region': 'country_name', 'value': 'total_ev_sales'}, inplace=True)

# Merging the datasets
merged_aqi_ev_df = pd.merge(air_quality_agg, ev_sales_total, on='country_name', how='inner')

# Plotting the interactive map
fig = px.scatter_geo(
    merged_aqi_ev_df,
    locations="country_name",
    locationmode="country names",
    size="total_ev_sales",
    color="avg_aqi",
    hover_name="country_name",
    size_max=50,
    title="Gemiddeld AQI vs verkoop elektrische voertuigen per land"
)

fig.update_layout(
    geo=dict(
        showframe=False,
        showcoastlines=False,
        projection_type='equirectangular'
    )
)

fig.show()

```

> *Figure 5: Een interactieve kaart die landen toont met hun gemiddelde luchtkwaliteitsindex (AQI) en het totale aantal verkochte elektrische voertuigen. Grotere stippen vertegenwoordigt een hogere verkoop van elektrische voertuigen, en de kleur geeft de gemiddelde AQI weer (paars voor goede luchtkwaliteit, geel voor slechte luchtkwaliteit).

Hoe groter de stippen in de kaart hoe meer verkoop van elektrische voertuigen er is, China is verreweg de grootste stip, wat betekent dat zij ook de meeste elektrische voertuigen verkopen. De stip van china is geel, dit betekent dat China een heel hoog gemiddeld aqi heeft. Een hoog aqi betekent een slechte luchtkwaliteit. Dit laat zien dat ondanks een heel groot aantal elektrische voertuigen, de luchtkwaliteit slecht blijft en er dus weinig tot geen impact is van elektische voertuigen op de luchtkwaliteit. 

### Meer CO2-uitstoot in de afgelopen jaren 

```python
import pandas as pd
import matplotlib.pyplot as plt

# Laad de CO2-uitstootgegevens
co2_emissions_df = pd.read_csv('co2_emissions_kt_by_country.csv')

# Filter de gegevens voor de jaren 2000 tot 2022
co2_emissions_filtered = co2_emissions_df[(co2_emissions_df['year'] >= 2000) & (co2_emissions_df['year'] <= 2022)]

# Agregeer de gegevens om de mondiale CO2-uitstoot per jaar te krijgen
global_co2_emissions = co2_emissions_filtered.groupby('year')['value'].sum().reset_index()
global_co2_emissions.rename(columns={'value': 'global_co2_emissions'}, inplace=True)

# Plot de mondiale CO2-uitstoot in de tijd
plt.figure(figsize=(14, 8))
plt.plot(global_co2_emissions['year'], global_co2_emissions['global_co2_emissions'], marker='o', linestyle='-', color='b')
plt.title('Mondiale CO2-uitstoot van 2000 tot 2022')
plt.xlabel('Jaar')
plt.ylabel('CO2-uitstoot (kt)')
plt.grid(True)
plt.show()


```

Door meer CO2-uitstoot wordt de luchtkwaliteit slechter. Ondanks de groei van de verkoop van elektrische voertuigen(figuur 1) blijft de CO2-uitstoot ook groeien. De groei van de verkoop van elektrische voertuigen heeft dus weinig tot geen impact op de luchtkwaliteit van landen.

> *Figure 6: X-as: Op deze as staat het jaartal, van 2000 tot en met 2022. Y-as: op deze as staat de CO2-uitstoot in kiloton keer 100 miljoen.

In figuur 6 zie je de groei van de CO2-uistoot in de afgelopen jaren. Het is duidelijk dat ondanks de groei van de verkoop van elektrische voertuigen, er nog steeds een groei is van de CO2-uitstoot. Deze grafiek laat zien dat elektrische voertuigen weinig tot geen impact hebben op de luchtkwaliteit van landen. 

### Verkoop van EV's in Verschillende Landen en Beperkte Impact op Luchtkwaliteit

Hoewel de verkoop van elektrische voertuigen in sommige landen toeneemt, blijft de impact op de luchtkwaliteit beperkt. Dit kan worden verklaard door het feit dat de verkoop van elektrische voertuigen geconcentreerd is in een beperkt aantal landen, terwijl andere landen nog steeds een zeer laag aantal elektrische voertuigen verkopen.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Laad de dataset
ev_sales_df = pd.read_csv('IEA-EV-dataEV salesCarsHistorical.csv')

# Verwijderen van niet-specifieke regio's
regions_to_exclude = ['World', 'Europe', 'EU27', 'Other Europe', 'Rest of the world']
ev_sales_df_filtered = ev_sales_df[~ev_sales_df['region'].isin(regions_to_exclude)]

# Aggregating the EV sales data by country
ev_sales_per_country = ev_sales_df_filtered.groupby('region')['value'].sum().reset_index()
ev_sales_per_country.rename(columns={'region': 'country_name', 'value': 'total_ev_sales'}, inplace=True)

# Sort the data by total EV sales for better visualization
ev_sales_per_country = ev_sales_per_country.sort_values(by='total_ev_sales', ascending=False)

# Plotten van de totale EV-verkoop per land
plt.figure(figsize=(14, 8))
plt.bar(ev_sales_per_country['country_name'], ev_sales_per_country['total_ev_sales'], color='b')
plt.title('Totale Verkoop van Elektrische Voertuigen per Land')
plt.xlabel('Land')
plt.ylabel('Totale EV Verkoop')
plt.xticks(rotation=90)
plt.grid(True)
plt.show()

```

> *Figure 7: X-as: Op deze as staan de landen. Y-as: op deze as staat de totale verkoop van elektrische voertuigen keer 10 miljoen.

De staafdiagram laat zien hoe de verkoop van elektrische voertuigen zich door de jaren heen in verschillende landen heeft ontwikkeld.
Het is duidelijk te zien dat bepaalde landen, zoals China en de Verenigde Staten, een aanzienlijk hogere verkoop van elektrische voertuigen hebben in vergelijking met andere landen. De geconcentreerde verkoop van elektrische voertuigen in een beperkt aantal landen betekent dat de impact op de wereldwijde luchtkwaliteit beperkt blijft, omdat veel landen nog steeds een zeer laag aantal EV's verkopen.
Deze visualisatie ondersteunt het argument dat, hoewel de verkoop van elektrische voertuigen stijgt in sommige landen, de wereldwijde impact op de luchtkwaliteit beperkt is omdat de verkoop geconcentreerd is in een klein aantal landen.
